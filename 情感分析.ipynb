```python
from snownlp import SnowNLP
import numpy as np 
```

```python
path='../aa5.txt'  ######读取comments
fn = open(path,encoding='utf-8')
string_data = fn.read()
fn.close()
```

```python
data_list=[]
ind_left=[]
ind_right=[]
for i in range(len(string_data)):
    if '{' ==string_data[i]:
        ind_left.append(i)
    if '}' ==string_data[i]:
        ind_right.append(i)
for j in range(len(ind_left)):
    data_list.append(string_data[ind_left[j]+13:ind_right[j]-7])
```

```python
good=0
medium=0
bad=0
for item in data_list:
    try:
        score=SnowNLP(item).sentiments
        if score >0.8:
            good+=1
        elif score >0.4:
            medium+=1
        else:
            bad+=1
            print (data_list.index(item))
    except:
        pass
print ('Number of good:', good)
print ('Number of medium:', medium)
print ('Number of bad:', bad)
```
